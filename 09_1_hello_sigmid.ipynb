```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.arange(-5.0,5.0,0.1)
```

#Step function (계단함수)

```
def step_function(x):
  return np.array(x>0, dtype=np.int)
```

x가 0보다 크면 true(1) 

```
y = step_function(x)
```

```
y.shape
```

```
plt.plot(x,y)
plt.ylim(-0.1,1.1) #-0.1 에서 1.1사이만 보기
plt.show()
```

#Sigmoid
A sigmoid function is a mathematical function having a characteristic "S"-shaped curve or sigmoid


$f(x) = \cfrac{1}{1+e^{-x}}$

frac -> 분수

c -> 글자크기 크게 

```
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```
x = np.arange(-5.0, 5.0, 0.1)
y2 = sigmoid(x)
```

```
plt.plot(x, y2)
plt.grid()  #그래프에 그리드
plt.ylim(-0.1, 1.1)
plt.show()
```

#일반적인 form

$f(x) = \cfrac{1}{1+e^{-c_{1}x}}$

```
def sigmoid2(x,c1=1):
  return 1/(1+np.exp(-c1*x))
```

```
y2 = sigmoid2(x,3)
```

```
plt.plot(x, y2)
plt.grid()
plt.ylim(-0.1, 1.1)
plt.show()
```

```
for i in range(10):  #10개의 그래프 그리기
  c1 = 0.2 * (i+1)  #(i+1) -> 0 방지
  y2 = sigmoid2(x,c1)
  plt.plot(x,y2)
plt.grid()
```

#보다 일반적인 form
$f(x) = \cfrac{1}{1+e^{-c_{1}(x-c_{2})}}$

```
def sigmoid3(x,c1=1,c2=0):  #c1,c2를 미리 입력해서 밑에서 인자값을 안받아도 괜찮다.
  return 1/(1+np.exp(-c1*(x-c2)))
```

```
y3 = sigmoid3(x) #인자를 안받는 값은 뒤로 넘겨야한다.
```

```
plt.plot(x, y3) #원래는 100개의 점으로 이어져있다.
plt.grid()
#plt.ylim(-0.1, 1.1)
plt.show()
```

```
for i in range(10):
  c2 = 0.3* i
  c1 = 3.5
  y3 = sigmoid3(x,c1,c2)
  plt.plot(x,y3)
plt.grid()
```
