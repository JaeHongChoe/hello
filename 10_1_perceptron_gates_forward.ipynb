```
import numpy as np
```

```
def AND(x1, x2):
 x = np.array([x1, x2])
 w = np.array([0.5, 0.5])
 b = -0.7
 tmp = w[0]*x[0] + w[1]*x[1] + b # tmp = np.sum(w*x) + b
 #print('tmp=',tmp)
 if tmp <= 0:
  return 0
 else:
  return 1
```

```
AND(1,0)
```

```
AND(1,1)
```

```
AND(0,1)
```

#NAND Gate

```
def NAND(x1, x2):
 x = np.array([x1, x2])
 w = np.array([-0.5, -0.5])
 b = 0.7
 tmp = w[0]*x[0] + w[1]*x[1] + b
 if tmp <= 0:
  return 0
 else:
  return 1
```

```
NAND(0,0)
```

```
NAND(0,1)
```

```
NAND(1,1)
```

```
NAND(1,0)
```

#OR Gate

```
def OR(x1, x2):
 x = np.array([x1, x2])
 w = np.array([0.5, 0.5])
 b = -0.2
 tmp = np.sum(w*x) + b
 if tmp <= 0:
  return 0
 else:
  return 1
```

```
OR(1,1)
```

```
OR(0,1)
```

```
OR(0,0)
```

#XOR Gate
모두 같으면 0을 출력하는 XOR은 단층 Percepteron으로 표현이 불가하다.

```
def XOR(x1, x2):
 s1 = NAND(x1, x2)
 s2 = OR(x1, x2)
 y = AND(s1, s2)
 return y
```

```
XOR(0,0)
```

```
XOR(0,1)
```
